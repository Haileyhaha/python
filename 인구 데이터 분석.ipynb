```python
# 우리동네와 인구구조가 가장 비슷한 지역은 어딜까요 ? 
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 기호 깨짐 방지

plt.title('신도림동과 화천면의 인구 비율')

result1=[]
result2=[]
data= csv.reader(open('age.csv'))

for row in data :
    if '신도림동' in row[0] : 
        for i in range(3, 104) :
            result1.append(int(row[i])/int(row[2]))
    if '화촌면' in row[0] : 
        for i in range(3, 104) :
            result2.append(int(row[i])/int(row[2]))
            
plt.plot(result1)
plt.plot(result2)
plt.show()
```

$$ s = \sum (기준지역의 해당 연령값 - 각지역의 해당 연령값)^{2} $$

```python
# 인구구조가 가장 비슷하다는 것은?

import csv
pivot =[]
data= csv.reader(open('age.csv'))
next(data)
next(data)
data = list(data)

name = input('궁금한 지역 이름을 입력해주세요 :  ')

for row in data :
    if name in row[0] : 
        for i in range(3, len(row)) :
          pivot.append(int(row[i]) / int(row[2]))
        break
mn= 10000
print(pivot)
for row in data:
    s = 0
    # 16 ~ 20 오차 계산
    for i in range(3, len(row)):
        row[i] = int(row[i])/int(row[2]) #비율로 바꿔주는 코드
        tmp = (row[i]-pivot[i-3]) **2 # 두값의 차이를 제곱 
        s = s+tmp #차이의 제곱을 누적해서더하기
    # 22 ~ 26 : 역대 최소 오차를 업데이트
    if s < mn and (name not in row[0]):
        result = []
        for i in range(3, len(row)):
            result.append(row[i])
        mn = s
        result_name = row[0]
plt.figure(dpi =300)
plt.style.use('ggplot')

plt.plot(pivot, label = name)
plt.plot(result, label = result_name)
plt.legend()
plt.show()
```
