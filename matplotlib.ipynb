## matplotlib 

```python
# 그래프를 그리고 싶을때 사용하는 라이브러리
# 1. 라이브러리 import
# 2. 적절한 그래프 선택
# 3. 그래프 보여주기

import matplotlib.pyplot as plt
# matplotlib의 하위 라이브러리 pyplot 
# alias -> plt
plt.plot([10,-4,15,10]) # plot 꺾은선 그래프
plt.show()
```

```python
# top 5 그래프
# 1. 꺾은선 그래프 plot
plt.plot([-10,5,20,13]) # x축은 알아서 1씩 증가
plt.show()
```

```python
# top 5 그래프
# 1. 꺾은선 그래프 plot
plt.plot([10,20,30,40],[-10,5,20,13]) # x축은 알아서 1씩 증가
plt.show()
```

```python
#2. 막대그래프 bar, barh
#plt.bar(x,y)

plt.bar([1,2,3,4],[-10,5,-3,8])
plt.show()
```

```python
#2-2 수평 막대 그래프 barh
plt.barh([1,2,3,4],[-10,5,-3,8])
plt.show()
```

```python
# 히스토그램 hist
data = [-10.2, -5.2 , -8.0 , -5.6, -7.8, -5.9 ,-5.0, -1.2, -3.4,-10.3,-11.3,-3.2]
# plt.hist(data[,bins=n]) # bins는 막대의 갯수 , []는 생략가능 default는 10개 
plt.hist(data,bins=30)
plt.show()
```

```python
# 박스 플랏 boxplot
# plt.boxplot(data)
data = [-10.2, -5.2 , -8.0 , -5.6, -7.8, -5.9 ,-5.0, -1.2, -3.4, -10.3, -11.3, -3.2, -2.9]
plt.boxplot(data)
plt.show() # 주황색은 중앙값 
```

```python
# 산점도 scatter

# plt.scatter(x,y)
# 기본 폰트로 '맑은 고딕' 사용
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 기호 깨짐 방지

plt.title('맑은 고딕 테스트')

high = [2.3, 3.0, 2.5, 1.0, 1.9, 3.3 , 4.5, 3.7, 5.6 ,4.4, 0.9]
low = [-10.2, -5.2 , -8.0 , -5.6, -7.8, -5.9 ,-5.0, -1.2, -3.4, -10.3, -11.3]
plt.scatter(high, low, c= high) # c는 컬러 설정
plt.jet()
plt.colorbar()
plt.show()
```

```python
# 기본 폰트로 '맑은 고딕' 사용
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 기호 깨짐 방지

plt.title('맑은 고딕 테스트')
# x축, y축 만들기 
plt.xlabel('최고기온')
plt.ylabel('최저기온')

#범례 만들기 # 그래프 색깔 바꾸기 
plt.plot(low,label='1월 최저기온', c='skyblue')
plt.plot(high,label='1월 최고기온',c ='hotpink')
plt.legend()

plt.show()
```
