```python
# 최고 기온 데이터 그래프로 표현해보기

import csv
data = csv.reader(open('seoultemp.csv'))
next(data) 
temp=[]
for row in data:
    if row[-1] != '':
        temp.append(float(row[-1]))

import matplotlib.pyplot as plt
plt.plot(temp)
plt.show()
plt.boxplot(temp)
plt.show()
plt.hist(temp, bins=100)
plt.show()
```

```python
import csv
data = csv.reader(open('seoultemp.csv'))
next(data) 
july =[]
august =[]
for row in data:
    if row[-1] != '' and row[-2] !='':
        if row[0].split('-')[1] == '07':
            july.append(float(row[-1])-float(row[-2]))
        if row[0].split('-')[1] == '08':
            august.append(float(row[-1])-float(row[-2]))

import matplotlib.pyplot as plt
plt.boxplot([july, august])
plt.show()
```

```python
#1년중 일교차가 가장 큰달은?
import csv
data = csv.reader(open('seoultemp.csv'))
next(data) 
x=[]

for i in range(12):
    x.append([])
print(x)

for row in data:
    if row[-1] != '' and row[-2] != '':
        x[int(row[0].split('-')[1]) - 1].append(float(row[-1])-float(row[-2]))

import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 기호 깨짐 방지

plt.title('월별 일교차')
plt.boxplot(x)
plt.show()
```
